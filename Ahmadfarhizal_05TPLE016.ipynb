```python
# Task 1
def get_sorted_companies(company_list):
    # Sort the companies by name in reverse order
    sorted_companies = sorted(company_list, key=lambda x: x['name'], reverse=True)
    
    # Create a new list with only the name of each company
    result = [{'name': company['name']} for company in sorted_companies]
    
    return result

# Get and print the sorted companies
sorted_company_list = get_sorted_companies(company_detail_list)
import json
print(json.dumps(sorted_company_list, indent=4))
```

```python
#task 3
def list_employees_by_domain(company_list, employee_list):
    domain_employees = {}
    
    # Create a mapping of company name to domain
    company_to_domain = {company['name']: company['domain'] for company in company_list}
    
    # Iterate through each employee
    for employee in employee_list:
        company = employee['company']
        full_name = employee['full_name']
        
        # Get the domain of the company
        domain = company_to_domain.get(company)
        
        if domain:
            if domain not in domain_employees:
                domain_employees[domain] = []
            domain_employees[domain].append(full_name)
    
    return domain_employees

# Get the list of employees by domain and print the result
employees_by_domain = list_employees_by_domain(company_detail_list, employee_detail_list)

# Print the output in the expected format
import json
print(json.dumps(employees_by_domain,indent=4))
```

```python
#task 4
def get_employees_with_country(company_list, employee_list):
    # Create a mapping of company name to country
    company_to_country = {company['name']: company['country'] for company in company_list}
    
    result = []
    
    # Iterate through each employee
    for employee in employee_list:
        company_name = employee['company']
        country = company_to_country.get(company_name)
        
        if country:
            result.append({
                'full_name': employee['full_name'],
                'company': company_name,
                'country': country
            })
    
    return result

# Get the list of employees with their company country and print the result
employees_with_country = get_employees_with_country(company_detail_list, employee_detail_list)

# Print the output in the expected format
import json
print(json.dumps(employees_with_country,indent=4))
```

```python
#task 5
def get_companies_with_employee_nationality(company_list, employee_list):
    company_nationality = {}
    
    # Iterate through each employee
    for employee in employee_list:
        company_name = employee['company']
        nationality = employee['nationality']
        
        # Initialize company entry if not present
        if company_name not in company_nationality:
            company_nationality[company_name] = {}
        
        # Count the nationality
        if nationality not in company_nationality[company_name]:
            company_nationality[company_name][nationality] = 0
        company_nationality[company_name][nationality] += 1
    
    # Convert to the desired output format
    result = []
    for company in company_list:
        name = company['name']
        if name in company_nationality:
            result.append({
                'company': name,
                'employee_nationality': company_nationality[name]
            })
    
    return result

# Get the companies with employee nationality and print the result
companies_with_nationality = get_companies_with_employee_nationality(company_detail_list, employee_detail_list)

# Print the output in the expected format
import json
print(json.dumps(companies_with_nationality, indent=4))
```
